```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create


# Read in the data

df = pd.read_csv(r'C:\Users\NOMAD\Google Drive\Google Drive\Data Analyst Portfolio\Movie Industry Python\projectfiles\movies.csv')
# df = dataframe
```

```python
# Let's look at the data

df.head()
```

```python
# Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# change data type of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Create correct year column
# released column is an object. must make it into a string value

df['yearcorrect'] = df['released'].astype(str).str[:4]

df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates, using 'company' column as an example but you can do for any column or multiple if you want to

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df
```

```python
# Budget high correlation, higher budget = more revenue
# Company will have a high correlation

# Hypothesis from data table
```

```python
# Scatter plot with budget vs gross revenue


plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn
# scatter_kws = plots and points, line_kws = average line of all points

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
